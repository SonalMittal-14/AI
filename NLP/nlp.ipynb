```python
import spacy
```

```python
nlp = spacy.load("en_core_web_sm") # we want to use english core package 
```

```python
doc = nlp("Dr. Strange loves pav bhaji of mumbai. hulk loves chat of delhi.")
```

```python
for sentence in doc.sents: # spacy calculated these sentences accurately 
    print(sentence)
```

```python
for sentence in doc.sents:
    for word in sentence:
        print(word)
```
